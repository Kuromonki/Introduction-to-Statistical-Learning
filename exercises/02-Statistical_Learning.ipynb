```python
# 9 e) There seems to be a negative correlation between mpg and displacement, horsepower and weight
    #  and a somewhat positive correlation between mpg and acceleration  
p = sns.PairGrid(auto, vars=['mpg', 'displacement', 'horsepower', 'weight', 'acceleration'], hue='cylinders')
p = p.map_diag(plt.hist, edgecolor="w", bins = 10)
p = p.map_lower(plt.scatter, edgecolor="w")
p = p.map_upper(plt.scatter, edgecolor="w")
p = p.add_legend();
```

```python
# origin is 1 for US, 2 for Europe and 3 for Japan/Asia
d = sns.PairGrid(auto, vars=['mpg', 'displacement', 'horsepower', 'weight', 'acceleration', 'year'], hue='origin')
d = d.map_diag(plt.hist, edgecolor="w")
d = d.map_lower(plt.scatter, edgecolor="w")
d = d.map_upper(plt.scatter, edgecolor="w")
d = d.add_legend();
```

```python
#9 f) displacement, horsepower, weight and number of cylinders seem to impact mpg negatively acceleration and year
    # on the other hand seem to have a positive impact on mpg. It can also be said that US cars have in general lower mpg 
    # values compared to European and Asian cars
```

# Exercise 10 - Boston Data Set

```python
# 10 a)
boston = pd.read_csv('../data/Boston.csv')
boston.head()
```

Columns description:

- `crim` per capita crime rate by town.
- `zn` proportion of residential land zoned for lots over 25,000 sq.ft.
- `indus` proportion of non-retail business acres per town.
- `chas` Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
- `nox` nitrogen oxides concentration (parts per 10 million).
- `rm` average number of rooms per dwelling.
- `age` proportion of owner-occupied units built prior to 1940.
- `dis` weighted mean of distances to five Boston employment centres.
- `rad` index of accessibility to radial highways.
- `tax` full-value property-tax rate per \$10,000.
- `ptratio` pupil-teacher ratio by town.
- `black` $1000(Bk-0.63)^2$ where Bk is the proportion of blacks by town.
- `lstat` lower status of the population (percent).
- `medv` median value of owner-occupied homes in \$1000s.

```python
# 10 b)
sns.pairplot(boston);
```

```python
# 10 c) housing age and distance to employment centers to be correlated to crime rate, age is positively correlated 
    # while distance is negatively correlated
```

```python
# 10 d)
boston[['crim', 'tax', 'ptratio']].describe()
```

```python
fig, axes = plt.subplots(ncols = 3, figsize = (16, 8))
sns.boxplot(data = boston['crim'], ax=axes[0]).set_title("Crime")
sns.boxplot(data = boston['tax'], ax=axes[1]).set_title("Tax")
sns.boxplot(data = boston['ptratio'], ax=axes[2]).set_title('Pupil-Teacher ratio')
plt.show()
```
